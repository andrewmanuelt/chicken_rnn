```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt

from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
```

```python
from keras.models import Sequential
from keras.layers import SimpleRNN
from keras.layers import Dense
```

```python
df = pd.read_csv('./chicken.csv')

df = df.set_index('index')

df.head()
```

```python
plt.plot(df['value'])

plt.xlabel('Date')
plt.ylabel('Value');
```

```python
window1 = df.shift(1)
window1 = window1.rename(columns={'value': 'L1'})

window1.head()
```

```python
window2 = df.shift(2)
window2 = window2.rename(columns={'value': 'L2'})

window2.head()
```

```python
window3 = df.shift(3)
window3 = window3.rename(columns={'value': 'L3'})

window3.head()
```

```python
window4 = df.shift(4)
window4 = window4.rename(columns={'value': 'L4'})

window4.head()
```

```python
df_plain = df.copy()

df_plain.rename(columns={'value': 'Data'}, inplace=True)

df_plain['L1'] = window1['L1'] 
df_plain['L2'] = window2['L2'] 
df_plain['L3'] = window3['L3']
df_plain['L4'] = window4['L4']

df_plain.head()
```

```python
df_log = np.log(df_plain)

df_log.head()
```

```python
index = df_log.reset_index()
index = index['index']
```

```python
df_log = df_log[4:]

df_log.head()
```

```python
maksimal = np.max(df_log)
minimal = np.min(df_log)
```

```python
def min_max_scaler(unit, df, column):
    return (unit - np.min(df[column])) / (np.max(df[column]) - np.min(df[column]))
```

```python
def un_min_max_scaler(unit):
    # 0.043963	
    return unit * (maksimal - minimal) + minimal
```

```python
df_log['Data'] = df_log['Data'].apply(lambda x: min_max_scaler(x, df_log, 'Data'))

df_log['L1'] = df_log['L1'].apply(lambda x: min_max_scaler(x, df_log, 'L1'))
df_log['L2'] = df_log['L2'].apply(lambda x: min_max_scaler(x, df_log, 'L2'))
df_log['L3'] = df_log['L3'].apply(lambda x: min_max_scaler(x, df_log, 'L3'))
df_log['L4'] = df_log['L4'].apply(lambda x: min_max_scaler(x, df_log, 'L4'))

df_log.head()
```

```python
train = df_log.iloc[:170]

train.tail()
```

```python
test = df_log.iloc[170:]

test
```

```python
X_train, y_train = train[['L1', 'L2', 'L3', 'L4']], train['Data']
```

```python
X_test, y_test = test[['L1', 'L2', 'L3', 'L4']], test['Data']
```

```python
model = Sequential([
    SimpleRNN(32, input_shape=(4, 1)),
    Dense(1)
])

model.compile(optimizer='adam', loss='mse')

history = model.fit(X_train, y_train, epochs=500)
```

```python
plt.plot(history.history['loss'])

plt.xlabel('Learning Rate (Epoch)')
plt.ylabel('Error Rate')
```

```python
y_pred_train = model.predict(X_train)
```

```python
df_reverse_train = pd.DataFrame(columns=['train', 'predict'])

df_reverse_train['train'] = y_train
df_reverse_train['predict'] = y_pred_train

df_reverse_train['train'] = df_reverse_train['train'].apply(lambda x: np.exp(un_min_max_scaler(x)))
df_reverse_train['predict'] = df_reverse_train['predict'].apply(lambda x: np.exp(un_min_max_scaler(x)))
```

```python
print('R squared:', r2_score(df_reverse_train['train'], df_reverse_train['predict']))
print('MSE:', mean_squared_error(df_reverse_train['train'], df_reverse_train['predict']))
print('MAE:', mean_absolute_error(df_reverse_train['train'], df_reverse_train['predict']))
print('MAPE:', mean_absolute_percentage_error(df_reverse_train['train'], df_reverse_train['predict']))
```

```python
plt.plot(df_reverse_train['train'], label="Actual", color='blue')
plt.plot(df_reverse_train['predict'], label="Predicted", color='red')

plt.legend()
plt.ylabel('Value')
plt.xlabel('Date')
```

```python
y_pred_test = model.predict(X_test)
```

```python
df_reverse_test = pd.DataFrame(columns=['test', 'predict'])

df_reverse_test['test'] = y_test
df_reverse_test['predict'] = y_pred_test

df_reverse_test['test'] = df_reverse_test['test'].apply(lambda x: np.exp(un_min_max_scaler(x)))
df_reverse_test['predict'] = df_reverse_test['predict'].apply(lambda x: np.exp(un_min_max_scaler(x)))
```

```python
print('R squared:', r2_score(df_reverse_test['test'], df_reverse_test['predict']))
print('MSE:', mean_squared_error(df_reverse_test['test'], df_reverse_test['predict']))
print('MAE:', mean_absolute_error(df_reverse_test['test'], df_reverse_test['predict']))
print('MAPE:', mean_absolute_percentage_error(df_reverse_test['test'], df_reverse_test['predict']))
```

```python
plt.plot(df_reverse_test['test'], label="Actual", color='blue')
plt.plot(df_reverse_test['predict'], label="Predicted", color='red')

plt.legend()
plt.ylabel('Value')
plt.xlabel('Date')
plt.xticks(rotation=90);
```
